```python
wv_wiki = api.load("glove-wiki-gigaword-100")
```

### 함수 정의

```python
def cmp_word( wlst1, wlst2, sim_std = 0.6):
    try:
        sim = wv_wiki.n_similarity(wlst1, wlst2)
        if sim >= sim_std:
            return sim
        else:
            return False
    except:
        return False
```

```python
def f_sdg_list(x):
    lst = str(x).split()
    return dict(words = lst[1:], size = len(lst[1:]), wt = float(lst[0]))
```

```python
df = pd.read_csv(fname_sdg_list, encoding='utf-8')
df_sdg = df.applymap(f_sdg_list)
```

```python
def f_set_score(x, sr_sdg):

    score = 0.0

    for dic in sr_sdg:
        lst = str(x).split()

        for idx in range(len(lst) - (dic['size'] - 1)):
            wlist1 = lst[idx:idx+dic['size']]
            sim = cmp_word(wlist1, dic['words'])
            if sim: score += (sim * dic['wt'])

    return score
```

```python
df_c['SDG01']  = 0.0; df_c['SDG02']  = 0.0; df_c['SDG03']  = 0.0;
df_c['SDG04']  = 0.0; df_c['SDG05']  = 0.0; df_c['SDG06']  = 0.0;
df_c['SDG07']  = 0.0; df_c['SDG08']  = 0.0; df_c['SDG09']  = 0.0;
df_c['SDG10']  = 0.0; df_c['SDG11']  = 0.0; df_c['SDG12']  = 0.0;
df_c['SDG13']  = 0.0; df_c['SDG14']  = 0.0; df_c['SDG15']  = 0.0;
df_c['SDG16']  = 0.0; df_c['SDG17']  = 0.0;
```

#### 계산

```python
%%time
print('FDNX>>> Run: calculate SDG score...')

for col in df_sdg:
    print('    >>> ', col)
    df_c[col] = df_c['Preprocessed'].apply(f_set_score, args = (df_sdg[col],) )
```
